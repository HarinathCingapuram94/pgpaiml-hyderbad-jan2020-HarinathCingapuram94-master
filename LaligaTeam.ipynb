```python
df['Points'] = df['Points'].astype(int)
points_df = df.sort_values(by='Points',ascending = False)
list(points_df['Team'].head(5)) # list of teams which came Top 5 in terms of points 
```

```python
import numpy as np
from pandas import Series,DataFrame
def Goal_diff_count(laliga): ## function with name “Goal_diff_count”
    l = []
    for i in laliga.index:
        Goal_diff_count = int(df.iloc[i,8]) - int(df.iloc[i,9])
        l.append(abs(Goal_diff_count))
    G_Diff_df = pd.DataFrame({'Team':list(laliga['Team']),'Goal_diff_count':l})
    return G_Diff_df

Goal_Differences_df = Goal_diff_count(df)
```
